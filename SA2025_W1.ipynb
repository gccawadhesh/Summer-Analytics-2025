### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
average_mpg = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
sns.lineplot(data=average_mpg, x='model_year', y='mpg', hue='origin', marker='o')
plt.title('Evolution of Average MPG Over the Years by Origin', fontsize=16)
plt.xlabel('Model Year', fontsize=14)
plt.ylabel('Average MPG', fontsize=14)
plt.legend(title='Origin')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
import pandas as pd

df = pd.DataFrame({
    'car_name': ['Car A', 'Car A', 'Car B', 'Car B', 'Car C'],
    'model_year': [2020, 2021, 2020, 2021, 2020],
    'mpg': [25, 26, 30, 31, 28]
})

print("Initial DataFrame:")
print(df.head())
print(df.columns)

car_counts = df.groupby(['car_name', 'model_year']).size().reset_index(name='count')

car_appearances = car_counts.groupby('car_name').agg(
    appearances=('count', 'sum'),
    model_years=('model_year', 'nunique')
).reset_index()

print("\nCar Appearances:")
print(car_appearances)

consistent_cars = car_appearances[car_appearances['model_years'] > 1]
print("\nConsistent Cars (more than one model year):")
print(consistent_cars)

mpg_std = df.groupby('car_name').agg(
    mpg_std=('mpg', 'std'),
    avg_mpg=('mpg', 'mean'),
    appearances=('car_name', 'count')
).reset_index()

print("\nMPG Statistics:")


print(mpg_std)

consistent_cars = consistent_cars.merge(mpg_std, on='car_name', how='left')
print("\nMerged DataFrame:")
print(consistent_cars)

consistent_cars = consistent_cars[consistent_cars['mpg_std'] < 1.0]
print("\nConsistent Cars with MPG Std < 1.0:")
print(consistent_cars)

if not consistent_cars.empty:
    result = consistent_cars.sort_values(by=['appearances_x', 'avg_mpg'], ascending=[False, False])
    final_result = result[['car_name', 'appearances_x', 'avg_mpg']]
    print("\nFinal Result:")
    print(final_result)
else:
    print("\nNo consistent cars found with MPG standard deviation < 1.0.")
```
